```python
import numpy as np
sales.dropna(subset = ["Name"], inplace = True)
for name in sales["Name"]:
    for franchise in franchise_names:
        if(franchise in name):
            sales["Name"].replace(
                to_replace = [name],
                value = franchise,
                inplace = True)
sales["User_Score"] = sales["User_Score"].apply(pd.to_numeric, errors = 'coerce')
```

```python
sales.info()
```

```python
sales.drop("NA_Sales", axis = 1, inplace = True)
sales.drop("EU_Sales", axis = 1, inplace = True)
sales.drop("JP_Sales", axis = 1, inplace = True)
sales.drop("Other_Sales", axis = 1, inplace = True)
sales.dropna(subset = ["Year_of_Release"], inplace = True)
num_attribs = ["Year_of_Release", "Critic_Score", "Critic_Count", "User_Score", "User_Count"]
cat_attribs = ["Name", "Platform", "Genre", "Publisher", "Developer", "Rating"]
sales.info()
```

```python
sales["Platform"] = sales["Platform"].astype(str)
sales["Name"] = sales["Name"].astype(str)
sales["Genre"] = sales["Genre"].astype(str)
sales["Publisher"] = sales["Publisher"].astype(str)
sales["Developer"] = sales["Developer"].astype(str)
sales["Rating"] = sales["Rating"].astype(str)
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler, Imputer, LabelBinarizer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline, FeatureUnion
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        return X[self.attribute_names].values
class MyLabelBinarizer(TransformerMixin):
    def __init__(self, *args, **kwargs):
        self.encoder = LabelBinarizer(*args, **kwargs)
    def fit(self, x, y=0):
        self.encoder.fit(x)
        return self
    def transform(self, x, y=0):
        return self.encoder.transform(x)
class MultiColumnLabelBinarizer(TransformerMixin):
    def _init_(self, columns = "None"):
        self.columns = columns
    def fit(self, X, y = None):
        return self
    def transform(self, X, y = None):
        output = X.copy()
        if self.columns is not None:
            for col in self.columns:
                output[col] = MyLabelBinarizer().fit_transform(output[col])
        else:
            for colname, col in output.iteritems():
                output[colname] = MyLabelBinarizer().fit_transform(col)
        return output
num_pipeline = Pipeline([
    ("selector", DataFrameSelector(num_attribs)),
    ("imputer", Imputer()),
    ("std_scaler", StandardScaler())
])
cat_pipeline = Pipeline([
    ("selector", DataFrameSelector(cat_attribs)),
    ("label_binarizer", MultiColumnLabelBinarizer())
])
full_pipeline = FeatureUnion(transformer_list = [
    ("num_pipeline", num_pipeline)
])
```

```python
sales = sales[22:]
sales
```

```python
train_set, test_set = train_test_split(sales, test_size = 0.2, random_state = 42)
```

```python
train_set_labels = train_set["Global_Sales"].copy()
train_set = train_set.drop("Global_Sales", axis = 1)
```

```python
train_set_prepared = full_pipeline.fit_transform(train_set)
```

```python
train_set_prepared
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.tree import DecisionTreeRegressor
forest_reg = RandomForestRegressor()
forest_reg.fit(train_set_prepared, train_set_labels)
forest_predictions = forest_reg.predict(train_set_prepared)
forest_mse = mean_squared_error(train_set_labels, forest_predictions)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {'n_estimators': [3, 10, 38], 'max_features': [2, 3, 4, 5, 6]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
]
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5, 
                          scoring = 'neg_mean_squared_error')
grid_search.fit(train_set_prepared, train_set_labels)
grid_search.best_params_
```

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print (np.sqrt(-mean_score), params)
```
